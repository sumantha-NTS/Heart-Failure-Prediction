```python
#Splitting the data into input and output
x = data.drop('DEATH_EVENT',axis=1)
y = data.DEATH_EVENT
print("Dimention of data is")
print("x : ",x.shape)
print("y : ",y.shape)
```

## Building the Logistic regression

```python
#importing necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, plot_confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.model_selection import GridSearchCV
```

```python
#splitting data into train and test
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
print("Dimention of data is")
print("x_train : ",x_train.shape);print("y_train : ",y_train.shape);
print("x_test : ",x_test.shape);print("y_test : ",y_test.shape)
```

```python
logis = LogisticRegression()
```

```python
#fitting the model to train dataset and predicting for test data
pred_logis = logis.fit(x_train,y_train).predict(x_test)
plot_confusion_matrix(logis,x_test,y_test)
plt.show()
print("REPORT :\n",classification_report(y_test,pred_logis))
```

```python
#plotting ROC curve for evaluating the model
fpr , tpr , thresholds = roc_curve(y_test,logis.predict_proba(x_test)[:,1])
roc_auc_score(y_test,pred_logis)
```

```python
sns.lineplot(fpr,tpr)
```

```python
#checking for Hyperparameters
logis.get_params().keys()
```

```python
#Selecting parameters using GridSearchCv
param_grid = {'C': [0.01, 0.1, 1, 10, 100],  
              'solver': ['newton-cg', 'lbfgs', 'liblinear'], 
              'penalty': ['l1','l2']} 
grid = GridSearchCV(logis,param_grid,cv=5,n_jobs=-1,verbose=1)
grid.fit(x_train,y_train)
```

```python
print("Score : ",grid.best_score_)
print("Parameters : ",grid.best_params_)
```

```python
#predicting for test data with grid searched hyperparameters
pred_logis = grid.predict(x_test)
plot_confusion_matrix(grid,x_test,y_test)
plt.show()
print("REPORT :\n",classification_report(y_test,pred_logis))
acc_logis = round(accuracy_score(y_test,pred_logis)*100,2)
```
